```python
import pandas as pd
import pandas_datareader as pdr
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df_mt = pdr.get_data_yahoo('MINDTREE.NS')
```

```python
df_mt
```

```python
df_mt['High'].plot(figsize=(12,4))
plt.show()
```

```python
# Xlimit & ylimit & coloring

df_mt['High'].plot(xlim = ['2020-01-01', '2021-09-01'], ylim=[0, 4000], c = 'red',ls = '--' ,figsize = (16,6))
plt.show()
```

```python
df_mt.index
```

```python
index = df_mt.loc['2020-01-01': '2021-09-01'].index
share_open = df_mt.loc['2020-01-01': '2021-09-01']['Open']
```

```python
figure, ax = plt.subplots(figsize = (10,4))
ax.plot(index, share_open)
plt.tight_layout()
figure.autofmt_xdate()
```

### Time Resampling

```python
df_mt.resample(rule = 'Y').max()
```

```python
df_mt.resample(rule = 'Q')['Close'].max().plot()
plt.show()
```

### Simple Moving Average

```python
df_mt['High'].rolling(window = 11).mean().head(12)
```

```python
df_mt['Close: 30 Days Rolling'] = df_mt['Close'].rolling(window = 30, min_periods=1).mean()
```

```python
df_mt[['Close: 30 Days Rolling', 'Close']].plot(figsize = (16, 4))
```

### Cumulative Moving Average - Expanding

```python
df_mt['High'].expanding(min_periods = 1).mean().plot(figsize=(16,5));
```

```python
df_mt['CMA'] = df_mt['High'].expanding(min_periods = 1).mean()
```

```python
df_mt[['High', 'CMA']].plot(figsize=(16,5))
```
